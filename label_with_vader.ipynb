```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from nltk.tokenize import sent_tokenize, RegexpTokenizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('stopwords')
nltk.download('punkt')
```

```python
df = pd.read_csv('/content/big_final_coindesk_scrape.csv')
```

```python
df = df.drop(0)
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
```

```python
stop_words = set(stopwords.words('english'))
cleaned_data = []
for i in range(len(df)):
    text = df.iloc[i]['Title']
    text = re.sub(r'[^\w\s]','',text)  # noktalama işaretlerini kaldır
    text = text.lower()  # küçük harflere çevir
    words = word_tokenize(text)  # kelimeleri tokenize et
    words = [word for word in words if not word in stop_words]  # stop words'leri kaldır
    cleaned_data.append(' '.join(words))  # kelimeleri birleştir
```

```python
df['New_Title'] = cleaned_data
```

```python
df = df.drop_duplicates(subset = ['Link'], keep = 'first')
df
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

```python
scores =[]
```

```python
for index, row in df.iterrows():
    sentence = row['Title']
    date = row['Date']
    link = row['Link']
    score = analyser.polarity_scores(sentence)
    scores.append((sentence, date, link, score))
```

```python
df_scores = pd.DataFrame(scores, columns=['Title', 'Date','Link','Score'])
```

```python
df_scores
```

```python
df_scores["Sentiment"] = ""
df_scores
```

```python
for index, row in df_scores.iterrows():
    score = row['Score']
    compound_score = score['compound']
    if compound_score >= 0.05 :
        row['Sentiment'] = "Positive"
    elif compound_score <= - 0.05 :
        row['Sentiment'] = "Negative"
    else :
       row['Sentiment'] = "Neutral"
```
